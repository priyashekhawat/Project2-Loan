```python
loan_status = df['loan_approved'].value_counts()

plt.pie(
    loan_status.values,
    labels=loan_status.index,
    autopct='%1.1f%%'
)
plt.title("Loan Approval Distribution")
plt.show()
```

```python
top = df.nlargest(10, 'loan_amount')

sns.barplot(
    x='loan_amount',
    y='income',
    data=top
)
plt.title("Top 10 Loan Amounts")
plt.show()
```

```python
sns.catplot(x= "loan_approved",y = "credit_score",kind= 'bar',data=df)
plt.show()
```

```python

sns.relplot(
    x='income',
    y='loan_amount',
    hue='loan_approved',
    data=df
)

plt.show()
```

```python
status_income = df.groupby('loan_approved')['income'].mean()

status_income.plot(kind='bar')
plt.title("Average Income by Loan Status")
plt.ylabel("Income")
plt.show()
```

```python
loan_status=df.groupby("loan_approved")[[
    "loan_amount",
    "income",
    "credit_score"
]].mean()

loan_status.plot(kind="bar")
plt.title("Loan Approval Analysis")
plt.xlabel("loan status")
plt.ylabel("average value")
plt.xticks(rotation=0)
plt.show()
```

```python
avg_price = df.groupby('name')['loan_approved'].mean().sort_values()

avg_price.plot(kind='bar')
plt.title("Average Price by Brand")
plt.ylabel("Price")
plt.show()
```
